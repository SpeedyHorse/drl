```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from glob import glob
import os
```

```python
def show_histogram(data, name):
    basename = os.path.basename(name)
    data = pd.Series(data.values.flatten())

    data.hist(grid=True, color="blue")
    plt.title(f"Histogram {basename}.csv std={data.std():.2f}")
    plt.show()
```

```python
files = glob('train/*.csv')
for file in files:
    data = pd.read_csv(file, header=None)
    show_histogram(data, file)
```
